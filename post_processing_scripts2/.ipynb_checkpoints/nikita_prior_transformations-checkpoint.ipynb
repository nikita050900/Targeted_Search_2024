```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%load_ext autoreload
#%load_ext line_profiler
%autoreload 2

import numpy as np

import matplotlib.pyplot as plt
```

```python
NN = 1_000_000
log10_Mc_min =  7.0 #M_sol 
log10_Mc_max = 10.0 #M_sol
log10_h_min = -18.0 
log10_h_max = -11.0
log10_f_min = -8.01 #Hz
log10_f_max = -8.00 #Hz
log10_dL_min = 3.00 #Mpc
log10_dL_max = 3.05 #Mpc

megaparsec = 3.086e+22 #meter
speed_of_light = 299792458.0 #m/s
T_sun = 1.327124400e20 / speed_of_light**3 #G*M_sun/c^3 = R_sun/c-->Sun time

#log Mc = 3/5 log h + const
const_0 = -3/5*np.log10(2) - 2/5*np.log10(np.pi) - 3/5*np.log10(T_sun) - 3/5*np.log10(speed_of_light/megaparsec)
const_min = -2/5*log10_f_max + 3/5*log10_dL_min + const_0
const_max = -2/5*log10_f_min + 3/5*log10_dL_max + const_0
print(const_0, const_min, const_max)

# fiducial_m = 1e9
# dist_UL_sky = 2 * (fiducial_m*T_sun)**(5/3) * (np.pi*(fmin+fmax)/2)**(2/3) / (10**UL_sky) 
# dist_UL_sky *= speed_of_light/megaparsec

log10_Mcs = np.random.uniform(log10_Mc_min,log10_Mc_max,NN)
log10_hs = np.random.uniform(log10_h_min,log10_h_max,NN)

plt.plot(log10_hs, log10_Mcs, ls='', marker='.')

#print(log10_Mcs-3/5*log10_hs)
dL_filter = np.where((log10_Mcs-3/5*log10_hs>const_min) & ((log10_Mcs-3/5*log10_hs<const_max)))[0]
#print(dL_filter)
plt.plot(log10_hs[dL_filter], log10_Mcs[dL_filter], ls='', marker='.')
```

```python
#bins = np.linspace(log10_Mc_min, log10_Mc_max, 30)
bins = np.linspace(max(log10_Mc_min, const_max+3/5*log10_h_min), min(log10_Mc_max, const_min+3/5*log10_h_max), 30)
plt.hist(log10_Mcs[dL_filter], bins=bins)
```

```python
NN = 1_000_000
log10_Mc_min =  7.0 #M_sol 
log10_Mc_max = 10.0 #M_sol
log10_h_min = -18.0 
log10_h_max = -11.0
log10_f_min = -8.5#-8.01 #Hz
log10_f_max = -7.0#-8.00 #Hz
log10_dL_min = 4.00 #Mpc
log10_dL_max = 4.05 #Mpc

megaparsec = 3.086e+22 #meter
speed_of_light = 299792458.0 #m/s
T_sun = 1.327124400e20 / speed_of_light**3 #G*M_sun/c^3 = R_sun/c-->Sun time

#log Mc = 3/5 log h + const
const_0 = -3/5*np.log10(2) - 2/5*np.log10(np.pi) - 3/5*np.log10(T_sun) - 3/5*np.log10(speed_of_light/megaparsec)
const_min = 3/5*log10_dL_min + const_0
const_max = 3/5*log10_dL_max + const_0
print(const_0, const_min, const_max)

# fiducial_m = 1e9
# dist_UL_sky = 2 * (fiducial_m*T_sun)**(5/3) * (np.pi*(fmin+fmax)/2)**(2/3) / (10**UL_sky) 
# dist_UL_sky *= speed_of_light/megaparsec

log10_Mcs = np.random.uniform(log10_Mc_min,log10_Mc_max,NN)
log10_hs = np.random.uniform(log10_h_min,log10_h_max,NN)
log10_fs = np.random.uniform(log10_f_min,log10_f_max,NN)

plt.plot(log10_hs, log10_Mcs, ls='', marker='.')

#print(log10_Mcs-3/5*log10_hs)
dL_filter = np.where((log10_Mcs-3/5*log10_hs+2/5*log10_fs>const_min) & ((log10_Mcs-3/5*log10_hs+2/5*log10_fs<const_max)))[0]
#print(dL_filter)
plt.plot(log10_hs[dL_filter], log10_Mcs[dL_filter], ls='', marker='.')
```

```python
#bins = np.linspace(log10_Mc_min, log10_Mc_max, 30)
bins = np.linspace(max(log10_Mc_min, const_max+3/5*log10_h_min-2/5*log10_f_min),
                   min(log10_Mc_max, const_min+3/5*log10_h_max-2/5*log10_f_max), 30)
print(bins)
plt.hist(log10_Mcs[dL_filter], bins=bins)
```

```python
NN = 10_000_000
log10_Mc_min =  7.0 #M_sol 
log10_Mc_max = 10.0 #M_sol
log10_h_min = -18.0 
log10_h_max = -11.0
log10_f_min = -8.50 #Hz
log10_f_max = -7.50 #Hz
log10_dL_min = 3.00 #Mpc
log10_dL_max = 3.05 #Mpc

megaparsec = 3.086e+22 #meter
speed_of_light = 299792458.0 #m/s
T_sun = 1.327124400e20 / speed_of_light**3 #G*M_sun/c^3 = R_sun/c-->Sun time

#log Mc = 3/5 log h + const
const_0 = -3/5*np.log10(2) - 2/5*np.log10(np.pi) - 3/5*np.log10(T_sun) - 3/5*np.log10(speed_of_light/megaparsec)
const_min = 3/5*log10_dL_min + const_0
const_max = 3/5*log10_dL_max + const_0
print(const_0, const_min, const_max)

# fiducial_m = 1e9
# dist_UL_sky = 2 * (fiducial_m*T_sun)**(5/3) * (np.pi*(fmin+fmax)/2)**(2/3) / (10**UL_sky) 
# dist_UL_sky *= speed_of_light/megaparsec

#log10_Mcs = np.random.uniform(log10_Mc_min,log10_Mc_max,NN)
log10_hs = np.random.uniform(log10_h_min,log10_h_max,NN)
log10_fs = np.random.uniform(log10_f_min,log10_f_max,NN)

log10_Mcs = np.log10(np.random.uniform(10**log10_Mc_min,10**log10_Mc_max,NN))
#log10_hs = np.log10(np.random.uniform(10**log10_h_min,10**log10_h_max,NN))
#log10_fs = np.log10(np.random.uniform(10**log10_f_min,10**log10_f_max,NN))

plt.plot(10**log10_hs, 10**log10_Mcs, ls='', marker='.')

#print(log10_Mcs-3/5*log10_hs)
dL_filter = np.where((log10_Mcs-3/5*log10_hs+2/5*log10_fs>const_min) & ((log10_Mcs-3/5*log10_hs+2/5*log10_fs<const_max)))[0]
#print(dL_filter)
plt.plot(10**log10_hs[dL_filter], 10**log10_Mcs[dL_filter], ls='', marker='.')
```

```python
#bins = np.linspace(0, 10**log10_Mc_max, 30)
bins = np.linspace(0.0,
                   10**min(log10_Mc_max, const_min+3/5*log10_h_max-2/5*log10_f_max), 30)
print(bins)
plt.hist(10**log10_Mcs[dL_filter], bins=bins)
```
